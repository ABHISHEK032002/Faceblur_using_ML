```python
import cv2
import numpy as np
from mtcnn.mtcnn import MTCNN
```

```python
video = cv2.VideoCapture(r'C:\Users\rana\Downloads\people_video_1.mp4')    #input video path
```

```python
detector = MTCNN()
```

```python
while True:
    ret, frame = video.read()
    if not ret:
        break
    faces = detector.detect_faces(frame)
    for face in faces:
        x, y, w, h = face['box']
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 2)
        face_img = frame[y:y+h, x:x+w]
        face_img = cv2.GaussianBlur(face_img, (99, 99), 0)
        frame[y:y+h, x:x+w] = face_img
    cv2.imshow('Video', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
video.release()
cv2.destroyAllWindows()
```
